# Sequential (Linear) Search with Unordered List

```python
def sequentialSearchUnorderedList(List, value):
    """
    Sequential Search for Unordered List
    """
    index = 0
    found = False
    while index < len(List) and not found:
        
        if List[index] == value:
            found = True
        else:
            index += 1
        
    return (found, index)
```

```python
print(sequentialSearchUnorderedList(List=[1,4,2,6,3,7,8], value=6))
```

```python
print(sequentialSearchUnorderedList(List=[1,4,2,6,3,7,8], value=5))
```

# Sequential (Linear) Search with Ordered List

```python
def sequentialSearchOrderedList(List, value):
    """
    Sequential Search for Ordered List
    """
    index = 0
    found = False
    stop = False
    
    while index < len(List) and not found and not stop:
        
        if List[index] == value:
            found = True
        else:
            if value < List[index]:
                stop = True
            else:
                index += 1
                
    return (found, index)
```

```python
print(sequentialSearchOrderedList(List=[2,3,5,7,8,12,15,100], value=7))
```

```python
print(sequentialSearchOrderedList(List=[2,3,5,7,8,12,15,100], value=6))
```

# 3- Jump Search

```python
 def jumpSearch(List,value):
        """
        Jump Search for sorted array
        """
        n = len(List)
        step = int(n**0.5)
        prev = 0
        
        while List[int(min(step,n)-1)] < value:
            prev = step
            step += int(n**0.5)
            
            if prev >= n:
                return -1
            
        while List[int(prev)] < value:
            prev += 1
            
            if prev == min(step, n):
                return -1
            
        if List[int(prev)] == value:
            return int(prev)
        
        return -1
```

```python
print(jumpSearch(List=[1,12,23,23,34,45,54,76,99], value=34))
```
