# 1.1- Binary Search Tree

```python
class Node(object):
    def __init__(self, key):
        """
        Constructor
        """
        self.value = key
        self.right = None
        self.left = None
```

```python
def insert(root, node):
    if root is None:
        root = node
    else:
        if root.value < node.value:
            if root.right is None:
                root.right = node
            else:
                insert(root.right, node)
        else:
            if root.left is None:
                root.left = node
            else:
                insert(root.left, node)
```

```python
root = Node(41)
insert(root=root, node=Node(65))
insert(root=root, node=Node(99))
insert(root=root, node=Node(50))
insert(root=root, node=Node(20))
insert(root=root, node=Node(11))
insert(root=root, node=Node(29))
insert(root=root, node=Node(51))
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.value)
        inorder(root.right)
    else:
        return None # base case
```

# Key Search

```python
class Node(object):
    def __init__(self, key):
        """
        Node Constructor
        """
        self.value = key
        self.right = None
        self.left = None
```

```python
def search(root, key):
    """
    Search key in tree
    """
    # base case
    if root is None or root.value == key:
        return root
    
    if root.value < key:
        return search(root.right, key)
    else:
        return search(root.left, key)
```

```python
def insert(root, node):
    if root is None:
        root = node
    else:
        if root.value < node.value:
            if root.right is None:
                root.right = node
            else:
                insert(root.right, node)
        else:
            if root.left is None:
                root.left = node
            else:
                insert(root.left, node)
```

```python
root = Node(41)
insert(root=root, node=Node(11))
insert(root=root, node=Node(20))
insert(root=root, node=Node(29))
insert(root=root, node=Node(65))
insert(root=root, node=Node(50))
insert(root=root, node=Node(99))
```
