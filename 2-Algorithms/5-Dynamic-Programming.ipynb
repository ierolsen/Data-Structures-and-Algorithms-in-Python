# Fibonacci Number with Recursion

```python
def rec_fibo(n):
    """
    Recursion function for Fibonacci Sequence
    """
    
    # base case
    if n <= 1:
        return n
    else:
        return (rec_fibo(n-1) + rec_fibo(n-2))
```

```python
[rec_fibo(i) for i in range(10)]
```

# Fibonacci Number with Dynamic Programming

```python
def dynamic_fibonacci(n):
    
    fibo_list = [0, 1]
    
    while len(fibo_list) < n+1:
        fibo_list.append(0) # expands list to "n" length
    
    # base case
    if n <=1:
        return 1
    else:
        if fibo_list[n-1] == 0:
            fibo_list[n-1] = dynamic_fibonacci(n-1)
        if fibo_list[n-2] == 0:
            fibo_list[n-2] = dynamic_fibonacci(n-2)
        
        fibo_list[n] = fibo_list[n-2] + fibo_list[n-1]
    return fibo_list[n]
```

```python
[dynamic_fibonacci(i) for i in range(10)]
```

# Catalan Numbers with Recursion

```python
def recursive_catalan(n):
    # base case
    if n <= 1:
        return 1
    
    res = 0
    
    for i in range(n):
        res += recursive_catalan(i) * recursive_catalan(n - i - 1)
        
    return res
```

```python
[recursive_catalan(i) for i in range(10)]
```

# Catalan Numbers with Dynamic Programming

```python
def dynamic_catalan(n):
    # base case
    if n <= 1:
        return 1
    # expands list to "n" length 
    catalan = [0 for i in range(n+1)]
    
    # initilaze first two numbers
    catalan[0], catalan[1] = 1, 1
    
    # fill the list
    for i in range(2, n+1):
        catalan[i] = 0
        for j in range(i):
            catalan[i] = catalan[i] + catalan[j]*catalan[i-j-1]
    return catalan[n]
```

```python
[dynamic_catalan(i) for i in range(10)]
```
