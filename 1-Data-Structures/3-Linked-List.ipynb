# 1- Linked List

```python
class Node(object):
    
    def __init__(self, value):
        """
        Create Node: Node has value and next point
        """
        self.value = value
        self.nextnode = None
        
    def setNextNode(self,node):
        """
        It sets what next node is
        """
        self.nextnode = node
        
    def getNextNode(self):
        """
        It returns next node
        """
        return self.nextnode
    
    def getNodeValue(self):
        """
        It returns value of node
        """
        return self.value
```

```python
ankara = Node("06")
aydin = Node("09")
izmir = Node("35")

# ankara => aydin
ankara.setNextNode(aydin)
print(ankara.getNextNode().getNodeValue())

# aydin => izmir
aydin.setNextNode(izmir)
print(aydin.getNextNode().getNodeValue())

# ankara => aydin => izmir
print(ankara.getNextNode().getNextNode().getNodeValue())
```

# 2- Doubly Linked List

```python
class DoublyLinkedListNode(object):
    
    def __init__(self, value):
        """
        Create a Node, noda has value, next pointer and previously pointer
        """
        self.value = value
        self.nextnode = None
        self.prevnode = None
        
    def setNextNode(self, node):
        """
        It sets what next node is
        """
        self.nextnode = node
        
    def setPrevNode(self, node):
        """
        It sets what previously node is
        """
        self.prevnode = node
        
    def getNextNode(self):
        """
        It returns next Node
        """
        return self.nextnode
    
    def getPrevNode(self):
        """
        It returns previously Node
        """
        return self.prevnode
    
    def getNodeValue(self):
        """
        It returns value which is inside of Node
        """
        return self.value
```

```python
# ankara, aydin, izmir
ankara = DoublyLinkedListNode("06")
aydin = DoublyLinkedListNode("09")
izmir = DoublyLinkedListNode("35")

# ankara => aydin
ankara.setNextNode(aydin)

# aydin => ankara
aydin.setPrevNode(ankara)

# ankara <=> aydin
print(aydin.getPrevNode().getNodeValue())
print(ankara.getNextNode().getNodeValue())

# aydin => izmir
aydin.setNextNode(izmir)
print(izmir.getPrevNode()) # None

# izmir <=> aydin
izmir.setPrevNode(aydin)
print(izmir.getPrevNode().getNodeValue())

# izmir => aydin => ankara
print(izmir.getPrevNode().getPrevNode().getNodeValue())
```

# Adding Node to Linked List

```python
# Node Class
class Node(object):
    
    def __init__(self, data):
        """
        Create Node
        """
        self.data = data
        self.next = None
```

```python
# Linked List Class

class LinkedList(object):
    
    def __init__(self):
        """
        Head Initializer
        """
        self.head = None
        
    def push(self, new_data):
        """
        Add Node to top (head) of Linked List
        """
        # Create a new node
        new_node = Node(new_data)
        
        # New node points new Node which is coming after itself
        new_node.next = self.head
        
        # head points new Node
        self.head = new_node
        
    def printLinkedList(self):
        """
        Print Linked List
        """
        temp = self.head
        print("Linked List: ")
        while temp:
            print(temp.data)
            temp = temp.next
```

```python
linked_list = LinkedList()

linked_list.push("tail")
linked_list.push(15)
linked_list.push(25)
linked_list.push("head")

linked_list.printLinkedList()
```

# Delete Node from Linked List

```python
# Linked List Class

class LinkedList(object):
    
    def __init__(self):
        """
        Head Initializer
        """
        self.head = None
        
    def push(self, new_data):
        """
        Add Node to top (head) of Linked List
        """
        # Create a new node
        new_node = Node(new_data)
        
        # New node points new Node which is coming after itself
        new_node.next = self.head
        
        # head points new Node
        self.head = new_node
        
    def insertAfter(self, prev_node, new_data):
        """
        Add new Node after the given Node (Previously Node)
        """
        # if prev Node:
        if prev_node is None:
            print("There is no Node !")
            return        
        
        # Create New Node
        new_node = Node(new_data)
        
        # new_node next => prev_node next
        new_node.next = prev_node.next
        
        # prev_node next => new_node
        prev_node.next = new_node
    
    def append(self, new_data):
        """
        Add Node end of Linked List
        """
        # Create a new Node
        new_node = Node(new_data)
        
        # if Linked List empty, make new Node that is added will be head
        if self.head is None:
            self.head = new_node
            return
        
        # start from head go to end of Linked List
        last = self.head
        while last.next:
            last = last.next
        
        # add new Node instead of Last.next = None
        last.next = new_node
        
    def deleteNode(self, key):
        """
        Delete Node for key
        """
        temp = self.head
        
        # if only there is a Node and it has the key
        if temp is not None:
            if temp.data == key:
                self.head = temp.next
                temp = None
                return            
        
        # find key parameter for deleting Node
        while temp is not None:
            if temp.data == key:
                break
                
            prev = temp
            temp = temp.next
        
        # delete Node
        prev.next = temp.next
        temp = None
        
    def printLinkedList(self):
        """
        Print Linked List
        """
        temp = self.head
        print("Linked List: ")
        while temp:
            print(temp.data)
            temp = temp.next
```

```python
linked_list = LinkedList()

linked_list.push("Turkey")
linked_list.push("Germany")
linked_list.push("Denmark")
linked_list.printLinkedList()
print()

# Delete Node from Linked List
linked_list.deleteNode("Turkey")
linked_list.printLinkedList()
```
