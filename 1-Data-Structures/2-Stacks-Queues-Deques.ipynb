# Stack

```python
class Stack(object):
    
    def __init__(self):
        """
        Initialize (Constructor)
        """
        self.items = []
        
    def isEmpty(self):
        """
        Checks if it is empty
        """
        return self.items == [] # boolean operation (True or False)
    
    def push(self, item):
        """
        Adds item to Stack
        """
        self.items.append(item)
        
    def pop(self):
        """
        Removes item from Stack
        """
        return self.items.pop()
    
    def top(self):
        """
        Shows last Item from Stack
        """
        return self.items[len(self.items)-1]
    
    def size(self):
        """
        Gives size of Stack
        """
        return len(self.items)
```

```python
stack = Stack()
print(stack.isEmpty())

stack.push("Erol")
print(stack.top())
stack.push(2021)
print(stack.top())
stack.push("05.01.2021")
print(stack.top())
print()
print(stack.size())
print()
stack.pop()
print(stack.top())
print()
print(stack.isEmpty())
print()
stack.pop()
stack.pop()
print(stack.isEmpty())
```

# Queues

```python
class Queue(object):
    
    def __init__(self):
        """
        Initiliazer (constructor)
        """
        self.items = []
        
    def isEmpty(self):
        """
        Checks if it is empty
        """
        return self.items == [] # boolean operations (True or False)
    
    def enqueue(self, item):
        """
        Adds item to Queue
        """
        self.items.insert(0, item) # insert adds head of that (0 => add to 0. index), append adds tail of that
        
    def dequeue(self):
        """
        Removes item from Queue
        """
        return self.items.pop()
    
    def size(self):
        """
        length of items (Queue)
        """
        return len(self.items)
```

```python
queue = Queue()

print(queue.isEmpty())

queue.enqueue("Erol")
queue.enqueue("Germany")
print("size",queue.size())

queue.dequeue()
print("size",queue.size())

queue.dequeue()
print("size",queue.size())

print(queue.isEmpty())
```

# Deque

```python
class Deque(object):
    
    def __init__(self):
        """
        Initializer (constructor)
        """
        self.items = []
        
    def isEmpty(self):
        """
        Checks if it is empty
        """
        return self.items == [] # boolean operations (True or False)
    
    def addFront(self,item):
        """
        Adds item to Deque in Front side
        """
        self.items.append(item)
        
    def addRear(self,item):
        """
        Adds item to Deque in Rear side
        """
        self.items.insert(0,item)
        
    def removeFront(self):
        """
        Removes item to Deque in Front side
        """
        return self.items.pop()
    
    def removeRear(self):
        """
        Removes item to Deque in Rear side
        """
        return self.items.pop(0) # 0 => First Index !
    
    def size(self):
        """
        length of items (Deque)
        """
        return len(self.items)
```

```python
deque = Deque()

print(deque.isEmpty())

deque.addFront("Deep")
deque.addRear("Learning")

print("size: ",deque.size())

deque.removeFront()
print("size: ",deque.size())
deque.removeRear()
print("size: ",deque.size())

print(deque.isEmpty())
```
